```python
import matplotlib
import numpy as np
import pandas as pd
```

```python
A = [0,0]
B = [3,0]
C = [1.5,3]
X = [0,3,1.5,0]
Y = [0,0,3,0]

points=pd.DataFrame({"X":X,"Y":Y})
```

```python
def show(labels):
    plt.figure(figsize=(8,8))
    plt.plot(points.X,points.Y)
    plt.title('With x and y labels')
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    for index,point in enumerate(labels):
        plt.scatter(points.X[index],points.Y[index])
        plt.annotate(point, # this is the text
                 (points.X[index],points.Y[index]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,6), # distance from text to points (x,y)
                 ha='center')
```

```python
labels=["A","B","C"]
```

```python
show(labels)
```

```python
# For triangles
def keepXContant(labels):
    _temp=labels[1];
    labels[1]=labels[2];
    labels[2]=_temp
    return labels
```

```python
lables=keepXContant(labels)
show(labels)
```

```python
# For triangles
def keepYContant(labels):
    _temp=labels[0];
    labels[0]=labels[2];
    labels[2]=_temp
    return labels
```

```python
lables=keepYContant(labels)
show(labels)
```

```python
# For triangles
def keepZContant(labels):
    _temp=labels[0];
    labels[0]=labels[1];
    labels[1]=_temp
    return labels
```

```python
lables=keepZContant(labels)
show(labels)
```
